```python
import matplotlib.pyplot as plt

def draw_chessboard():
    fig, ax = plt.subplots()

    # Create a chessboard with 8x8 squares
    for i in range(8):
        for j in range(8):
            square_color = 'yellow' if (i + j) % 2 == 0 else 'purple'
            ax.add_patch(plt.Rectangle((i, j), 1, 1, facecolor=square_color))

    # Add pieces to the board
    ax.text(3.5, 3.5, 'K', fontsize=24, ha='center', va='center', color='white')
    ax.text(7.5, 7.5, 'Q', fontsize=24, ha='center', va='center', color='black')
    ax.text(7.5, 0.5, 'R', fontsize=24, ha='center', va='center', color='black')

    # Set axis limits and remove axis labels
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    ax.set_xticks([])
    ax.set_yticks([])

    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 grid for the four squares
fig, axs = plt.subplots(2, 2, figsize=(6, 6))

# Define a colormap for yellow and purple
cmap = plt.get_cmap('YlGnBu')

# Create matrices ranging from -1 to 1
matrix1 = np.random.uniform(-1, 1, size=(10, 10))
matrix2 = np.random.uniform(-1, 1, size=(10, 10))
matrix3 = np.random.uniform(-1, 1, size=(10, 10))
matrix4 = np.random.uniform(-1, 1, size=(10, 10))

# Plot the four squares
axs[0, 0].imshow(matrix1, cmap=cmap)
axs[0, 0].set_title("Yellow-Purple Square 1")

axs[0, 1].imshow(matrix2, cmap=cmap)
axs[0, 1].set_title("Yellow-Purple Square 2")

axs[1, 0].imshow(matrix3, cmap=cmap)
axs[1, 0].set_title("Yellow-Purple Square 3")

axs[1, 1].imshow(matrix4, cmap=cmap)
axs[1, 1].set_title("Yellow-Purple Square 4")

# Remove axis labels and ticks
for ax in axs.flat:
    ax.set_xticks([])
    ax.set_yticks([])

# Adjust spacing between subplots
plt.tight_layout()

plt.show()






```
