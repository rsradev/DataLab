````python
import numpy as np
import matplotlib.pyplot as plt

# Генерираме стойности за a
a = np.linspace(-10, 10)

# Изчисляваме b за всяко уравнение
b1 = 10 - a  # от a + b = 10
b2 = (24 - a) / 2  # от a + 2b = 12

# Рисуваме графиките
plt.figure(figsize=(8,6))
plt.plot(a, b1, label=r'$a + b = 10$')
plt.plot(a, b2, label=r'$a + 2b = 12$')

# Добавяме етикети и заглавие
plt.title('Графики на системата от уравнения')
plt.xlabel('a')
plt.ylabel('b')

# Добавяме легенда
plt.legend()

# Показваме графиката
plt.grid(True)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.show()
````
