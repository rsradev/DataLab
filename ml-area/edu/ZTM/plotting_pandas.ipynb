```python
heart_disease = pd.read_csv('./data/heart-disease.csv')
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(
    figsize=(15, 15),
    subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
#over_50 = heart_disease[heart_disease['target'] != 0]
```

```python
over_50
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(
    kind='scatter',
    x='age',
    y='chol',
    c='target',
    ax=ax
)

ax.set_xlim([45,100])
```

```python
over_50.dtypes
```

```python
over_50.head()
```

```python
plt.style.use('seaborn-v0_8')

fig, (ax0, ax1) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10,10),
    sharex=True
)
scatter=ax0.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
    cmap='winter'
)

scatter=ax1.scatter(
    x=over_50['age'],
    y=over_50['thalach'],
    c=over_50['target'],
    cmap='winter'
)
ax0.set_xlim([50, 80])
ax1.set_xlim([50, 80])
ax0.set(
    title='Heart Disease and Cholesterol Levels',
    xlabel='Age', # abs
    ylabel='Cholesterol'   
)

# add a legend
ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(over_50['chol'].mean(), linestyle='--');

ax1.set(
    title='Heart Disease and Max Heart Rate',
    xlabel='Age', # abs
    ylabel='Max Heart Rate'   
)
# add a legend

ax1.legend(*scatter.legend_elements(), title='Target')

ax1.axhline(over_50['thalach'].mean(), linestyle='--');

fig.suptitle(
    'Heart Disease Analysos',
    fontsize=16,
    fontweight='bold'
);
```
